## Observations and Insights 
Three observations:
    * The number of mice per drug ranged from aobut 150 to 230, and the ratio of males to females was very close to 50/50.
    * My average tumor volume data for the 4 best drug treatements did not have any outliers.
    * Mice treated with Capomulin and Ramicane finished the trial with smaller tumor sizes compared to those on Ceftamin and Infubinol.
    * Mouse weights had a strong positive correlation to tumor size.  The larger the mouse, the larger the tumor.

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.

mean = cleancombineddata.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
median = cleancombineddata.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
variance = cleancombineddata.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
stdev = cleancombineddata.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
sem = cleancombineddata.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()
summarystats = pd.DataFrame({"Mean": mean, "Median": median, "Variance": variance, "Standard Deviation": stdev, 
                          "SEM": sem})
summarystats
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
mouse_weights_cap = average_tumor_vol_cap["Weight (g)"].tolist()
mouse_avgTvol_cap = average_tumor_vol_cap["Average Tumor Volume (mm3)"].tolist()
r=st.pearsonr(mouse_weights_cap,mouse_avgTvol_cap)
print(f"The correlation between mouse weight and average tumor volume for Capomulin is {r[0]}")
```

```python
x_val = average_tumor_vol_cap["Weight (g)"]
y_val = average_tumor_vol_cap["Average Tumor Volume (mm3)"]
plt.scatter(x_val, y_val)

slope, intercept, r_value, p_value, std_err = st.linregress(x_val, y_val)
y = slope * x_val + intercept
plt.plot(x_val, y, color="r")
plt.title("Mouse Weight vs. Average Tumor Volume")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
